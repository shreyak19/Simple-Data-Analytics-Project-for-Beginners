# Display number of students in differetn boards -like central/other 10th

```python
a = data.groupby('ssc_b')['sl_no'].count().reset_index()
```

```python
a
```

```python
label = a['ssc_b']
values = a['sl_no']
```

```python
plt.axis("equal")
plt.pie(values,labels = label,radius=1.5,autopct='%0.2f%%')
plt.show()
```

# Display number of students in each steam ie: Science/commerce /arts 12th

# Count of male and female students

# Count of placed and not placed male and female

```python
d = data.groupby(['gender','status'])['sl_no'].count().reset_index()
```

```python
d
```

```python
label = d['gender']
values = d['sl_no']
```

```python
plt.axis("equal")
colors = ['y','m']
plt.pie(values,labels = label,radius=1.5,colors=colors,autopct='%0.2f%%')
plt.show()
```
